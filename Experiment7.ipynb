##HAC

Hierarchical clustering algorithms actually fall into 2 categories:


* Agglomerative (HAC): bottom-up, first assigns every example to its own cluster, and iteratively merges the closest clusters to create a hierarchical tree.


* Divisive (HDC): top-down, first groups all examples into one cluster and then iteratively divides the cluster into a hierarchical tree.

## Agglomerate two items at a time into a new cluster.


1.   Complete linkage: similarity of the farthest pair.
2.   Single-linkage: similarity of the closest pair.
3. Group average: similarity between groups.
4. Centroid similarity: each iteration merges the clusters with the most similar central point.

```python
myfields=['acq','corn','crude', 'earn', 'grain', 'interest', 'money-fx', 'ship', 'trade', 'wheat']
```

```python
def solution(data,idata, n):
  for x in set(data):
    if data.count(x) > n:
      temp=[]
      for i in range(len(data)):
        if x==data[i]:
          temp.append(i)
      for j in sorted(temp, reverse=True):
      #for j in range(len(temp)):
        del data[j]
        del idata[j]
  return data,idata
```

```python
files_ids=[]
label_ids=[]
id=0
for field in myfields:
  temp=reuters.fileids(field)
  temp_id=[id]*len(temp)
  files_ids.extend(temp)
  label_ids.extend(temp_id)
  id+=1
print(len(files_ids))
files_ids1,label_ids1= solution(files_ids,label_ids,1)
```
